 1. Viết chương trình thực hiện thuật toán

 3. So sánh với thuật toán Brute Force


# 1. Chương trình thực hiện thuật toán

```python
from random import randint

def quicksort(Arr):
    if len(Arr) < 2:
        return Arr
    low, same, high = [], [], []
    Val_mid = Arr[len(Arr) // 2][0]

    for item in Arr:
        if item[0] < Val_mid:
            low.append(item)
        elif item[0] == Val_mid:
            same.append(item)
        elif item[0] > Val_mid:
            high.append(item)
    return quicksort(low) + same + quicksort(high)

n = int(input())
A = []
[A.append(int(i)) for i in input().split()]
B = []
for i in range(0, n, 1):
  B.append([A[i], i])
B = quicksort(B)
print(B[n-1][1])
```

# 2. Giải recurrence relation của bài toán

T(n) = 2T(n/2) + n

T(n) = O(nlogn)
